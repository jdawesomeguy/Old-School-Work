```
import random
import string
import numpy as np
import copy
import math
```

Define Constants:

n: population size, default to 100

top: number of parents to be selected, defauylt to n/2




```
n = 100
top = n/2
letters = string.ascii_letters+" "+"!"+"?"+"."+","+"$"+"'"
```

```
#generate n copies of random strings with the length of "length"
def generate_population (length, n):
  parents = []

  for i in range(n):
    parent = ""
    for j in range(length):
      parent += random.choice(letters)
    parents.append(parent)

  return parents
```

```
#the number of characters that are in the same location in "target" and in "generated"
def fit_func(target, generated):
  fit = 0

  for i in range(len(target)):
    if target[i] == generated[i]:
      fit += 1

  return fit

```

```
fit_func("abc", "abe")
```

```
#select the top half of the population
def select_parents (target, parents):

  found = False

  selected = []

  parents = sorted(parents, key=lambda x: fit_func(target, x), reverse=True)
  parents = parents[:int(top)]

  for i in range(0, math.ceil(len(parents)/2)):
    selected.append(parents[i])
    if parents[i] == target:
      found = True
      break


  return selected, found

```

```
#randomly change something in your solution
def mutation(gene, mutation_n):

  for i in range(mutation_n):
    index = random.randint(0,len(gene)-1)
    gene = gene[:index]+random.choice(letters)+gene[index+1:]

  return gene

mutation("abcdefghi", 1)
```

Main loop starts here:

```
target = input ("Input: ")
print ("target: ", target)
length = len(target)
print (length)

parents = generate_population(length,n)
print (parents)

found = False
iteration_num = 0

while found == False:
  iteration_num +=1
  print ("iteration ", iteration_num)
  selected, found = select_parents (target, parents)
  print (selected)


  mutated = []
  mutation_n = 1
  for gene in selected:
    if random.random()>0.8:
      mutated.append(mutation(gene, mutation_n))
    else:
      mutated.append(gene)

  parents = selected+mutated


```
